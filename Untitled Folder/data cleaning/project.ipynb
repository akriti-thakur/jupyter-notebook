OVERALL SALES TREND

```python
# file['order_date'].min()
file['order_date'].max()
```

```python
file['month_year']= file['order_date'].apply(lambda x: x.strftime('%y-%m'))
```

```python
file
```

```python
trend=file.groupby(['month_year']).sum(numeric_only=True)['sales'].reset_index()
trend
```

```python
plt.figure(figsize=(15,5))
plt.plot(trend['month_year'],trend['sales'],color='red')
plt.xticks(rotation='vertical',size=10)
plt.yticks(rotation='vertical',size=6)
plt.show()
```

WHAT ARE THE TOP 10 PRODUCTS

```python
file['product_name']
```

```python
top_pro= file.groupby(['product_name']).sum(numeric_only='True')['sales']
top_pro=p.DataFrame(top_pro)
top_pro

#or
# top_pro=p.DataFrame( file.groupby(['product_name']).sum(numeric_only='True')['sales'])

```

```python
top_pro= top_pro.sort_values('sales',ascending=False)
top_pro
```
