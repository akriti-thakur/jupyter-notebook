```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as s
from re import match
import re
pd.set_option('display.max_columns', None)
```

```python
def downlo(url,filename):
     response= requests.get(url)
     if response.status_code==200:
         with open (filename,"wb") as f:
             f.write(response.content)
             
path="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0103EN-SkillsNetwork/labs/Module%202/recipes.csv"
downlo(path,"rec.csv")
```

```python
recipes.rename(columns={'country': 'cuisine'}, inplace=True)
recipes
```

```python
recipes['cuisine'] = recipes['cuisine'].str.lower()
recipes
```

```python
recipes_count= recipes['cuisine'].value_counts()

recipes_count

filter = recipes_count > 50
to_not_keep=recipes_count[~filter]
to_keep=recipes_count[filter]
to_keep_list=to_keep.index.to_list()
to_not_keep_list=to_not_keep.index.to_list()
```

```python
rows_before= recipes.shape[0]
print(f'number of rows in the original dataframe {rows_before}')



print(f"number of rows to keep{to_keep_list}")
print(f"{to_not_keep_list} rows removed!")
```
