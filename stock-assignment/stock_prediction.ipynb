```python
from telethon import TelegramClient

import pandas as pd
from telethon.sessions import MemorySession  # Use in-memory session to avoid SQLite

api_id = 22513475
api_hash = 'a3396d88d40e97287c74ef6c7ffd1cfc'
phone_number = '+916230606551'
channel = ["@marketmasters007"]  # Corrected the channel definition

async def main():
    async with TelegramClient(MemorySession(), api_id, api_hash) as client:  # Use MemorySession
        print("Client started")
        await client.start(phone_number)  # Start the client
        
        df = pd.DataFrame()
        for chat_name in channel:
            chat = await client.get_entity(chat_name)  # Corrected to await
            
            # Fetch messages
            messages_found = 0
            async for message in client.iter_messages(chat, reverse=True, limit=1000):  # Added limit for example
                print(f"Message found: {message.text}")
                data = {"group": chat_name, "text": message.text, "date": message.date}  # Added date
                temp_df = pd.DataFrame([data])  # Corrected DataFrame creation
                df = pd.concat([df, temp_df], ignore_index=True)  # Using pd.concat for better performance
                messages_found += 1

            print(f"Total messages retrieved: {messages_found}")

        # Check DataFrame content before saving
        print(df.head())

        # Remove timezone information
        if not df.empty:  # Check if DataFrame is not empty
            df['date'] = pd.to_datetime(df['date']).dt.tz_localize(None)
        
        # Save to CSV
        df.to_csv("data1.csv", index=False)  # Save to a different location

        # Confirm file write operation
        print("Data saved to CSV file.")

# Run the main function
import asyncio
asyncio.run(main())



```
